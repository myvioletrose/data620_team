## Draw bipartite graph 

```python
fig = plt.figure(figsize = (18, 10), dpi = 100, facecolor = 'w', edgecolor = 'k')

# set positions, layout for all nodes
pos = nx.fruchterman_reingold_layout(G) 

# get the node name as the key and the label as its value 
women_labels = {}    
for node in G.nodes():
    if node in women_nodes:        
        women_labels[node] = node              
        
event_labels = {}    
for node in G.nodes():
    if node in event_nodes:      
        event_labels[node] = node

# draw graph with no label
nx.draw(G, pos, with_labels = False, node_color = "lime")

# draw nodes, change font size, color
nx.draw_networkx_labels(G, 
                        pos, 
                        nodelist = women_nodes, 
                        labels = women_labels,                          
                        font_size = 16,
                        font_color = "red")

nx.draw_networkx_labels(G,
                        pos,
                        nodelist = event_nodes,
                        labels = event_labels,                        
                        font_size = 24,
                        font_color = "black")

# edges
nx.draw_networkx_edges(G, pos, width = 1.0, alpha = 0.5)
```

## Centrality measure

```python
women_nodes_with_centrality = [node for node in nx.bipartite.degree_centrality(G, women_nodes).items() if len(node[0]) >=10]

pd.Series(data = dict(women_nodes_with_centrality)).sort_values(ascending = False)
```

```python
event_nodes_with_centrality = [node for node in nx.bipartite.degree_centrality(G, event_nodes).items() if len(node[0]) <=3]

pd.Series(data = dict(event_nodes_with_centrality)).sort_values(ascending = False)
```

Take a quick look from above measures, it seems that Evelyn Jefferson, Nora Fayette, Theresa Anderson are the most sociable people and events 7, 8 and 9 seem the most important as they connect most people/attendees. 

## Creata a summary df

Caution when reading below summary table, it is a mix of women and events, i.e. that's a many-to-many relationship. For example, Evelyn Jefferson has the degree centrality of 0.571429, and she attended 8 separate events. One of them is E8 which has the degree centrality of 0.777778 and 14 attendees. 

```python
women_centralility = pd.Series(data = dict(women_nodes_with_centrality)).sort_values(ascending = False).to_frame()
women_centralility.reset_index(inplace = True)
women_centralility.columns = ['women', 'women_degree_centrality']

event_centralility = pd.Series(data = dict(event_nodes_with_centrality)).sort_values(ascending = False).to_frame()
event_centralility.reset_index(inplace = True)
event_centralility.columns = ['event', 'event_degree_centrality']

df = pd.DataFrame(list(G.edges(data = True)), columns = ['women', 'event', 'attribute'])\
.drop(columns = ['attribute'])\
.merge(women_centralility, on = "women")\
.merge(event_centralility, on = "event")

w = pd.DataFrame({'event_attended' : df.groupby("women").size()})
e = pd.DataFrame({'event_attendees' : df.groupby("event").size()})

df = df\
.merge(w, on = "women")\
.merge(e, on = "event")
```

```python
pd.set_option('display.max_rows', df.shape[0]+1)
df = df.sort_values(by = ["women_degree_centrality", "women", "event_degree_centrality"], ascending = [False, True, False])\
.reset_index(drop = True)
df
```

## Projections on "Women" (as primary set of nodes)

Let's use "women" as a primary set and do projections on it.

```python
women_graph = nx.bipartite.projection.projected_graph(G, women_nodes)

for n, d in women_graph.nodes(data = True):
    women_graph.node[n]['neighbors_count'] = len(list(women_graph.neighbors(n)))

list(women_graph.node(data=True))
```

```python
fig = plt.figure(figsize = (18, 10), dpi = 100, facecolor = 'w', edgecolor = 'k')
nx.draw(women_graph, 
        node_color = "lime",
        font_size = 12,
        node_order = 'neighbors_count', 
        layout = 'spring',
        with_labels = True)
```

Above graph is a bit hard to read, it seems many of them are interconnected to each other (as most have large number of neighbors). What would happen if we break the graph by removing the three women?

```python
women_graph.remove_nodes_from(["Evelyn Jefferson", "Nora Fayette", "Theresa Anderson"])
list(women_graph.node(data = True))
```

```python
fig = plt.figure(figsize = (18, 10), dpi = 100, facecolor = 'w', edgecolor = 'k')
nx.draw(women_graph, 
        node_color = "pink",
        font_size = 12,
        node_order = 'neighbors_count', 
        layout = 'random',
        with_labels = True)
```

## Cliques

```python
women_graph = nx.bipartite.projection.projected_graph(G, women_nodes)
    
women_cliques = list(nx.find_cliques(women_graph))

print('there are {x} cliques based on women in this network'.format(x = len(women_cliques)))

women_cliques
```

```python
fig = plt.figure(figsize = (18, 10), dpi = 100, facecolor = 'w', edgecolor = 'k')
women_graph_projected = nx.bipartite.projection.projected_graph(G, women_nodes)

plt.subplot(131)
nx.draw(women_graph_projected.subgraph(women_cliques[0]), with_labels = True, node_color = "gold", font_color = "red")

plt.subplot(132)
nx.draw(women_graph_projected.subgraph(women_cliques[1]), with_labels = True, node_color = "blue", font_color = "red")

plt.subplot(133)
nx.draw(women_graph_projected.subgraph(women_cliques[2]), with_labels = True, node_color = "grey", font_color = "red")
```

```python
event_graph = nx.bipartite.projection.projected_graph(G, event_nodes)
    
event_cliques = list(nx.find_cliques(event_graph))

print('there are {x} cliques based on events in this network'.format(x = len(event_cliques)))

event_cliques
```

```python
fig = plt.figure(figsize = (18, 10), dpi = 100, facecolor = 'w', edgecolor = 'k')
event_graph_projected = nx.bipartite.projection.projected_graph(G, event_nodes)

plt.subplot(121)
nx.draw(event_graph_projected.subgraph(event_cliques[0]), with_labels = True, node_color = "lime", font_color = "black")

plt.subplot(122)
nx.draw(event_graph_projected.subgraph(event_cliques[1]), with_labels = True, node_color = "yellow", font_color = "black")
```
